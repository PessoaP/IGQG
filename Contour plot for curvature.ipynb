```python
from IGQG import *
from numpy import array,arange,meshgrid,concatenate, ones

from matplotlib import pyplot as plt
```

This notebook makes the countour graphs for curvature of both fermions and bosons presented in Fig. 3 and Fig. 6 in the manuscript, respectively.

The commented cells [4] and [9] obtain the graphs for $\eta=2$

```python
eta = power(2,-1)
```

Fermions


```python
x = arange(.01,1,.01)
```

```python
#eta=2
```

```python
aarray = array([A(eta,-xx) for xx in x])
barray = array([B(eta,-xx) for xx in x])
```

```python
Rad = barray*pow(aarray,-2)
betaf = 1/arange(10,100,2) #l

X,Y = meshgrid(x,betaf)
Z = .5*Rad*(Y**(eta+1))
```

```python
fig, ax = plt.subplots()
CF = ax.contourf(X,Y,Z)
CS = ax.contour(X,Y,Z,colors='k')

ax.set_xlabel(r'$\xi$',fontsize=16)
ax.set_ylabel(r'$\beta$',fontsize=16)
ax.set_title(r'$R$ (fermions) for $\eta =$ {} '.format(eta))

plt.colorbar(CF)
plt.savefig('Rfermions-eta{}.png'.format(eta), bbox_inches = "tight",dpi=900)

plt.show()
```

Bosons

```python
x=concatenate([arange(.01,.9,.01),arange(.9,.99,.001),arange(.99,.999,.0002)])
beta = concatenate([arange(0,.03,.001),arange(.03,.2005,.005)])
```

```python
#eta=2
#beta = concatenate([beta,arange(.2,.601,.01)])
```

```python
aarray = array([A(eta,xx) for xx in x])
barray = array([B(eta,xx) for xx in x])

acarray = array([Ac(eta,xx) for xx in x])
bcarray = array([Bc(eta,xx) for xx in x])
```

```python
X,Y = meshgrid(x,beta)
ide,Ye = ones(Y.shape),Y**(eta+1)

num =  -Ye*(barray+Ye*bcarray)
den = 2*((aarray+Ye*acarray)**2)
Z = num/den
```

```python
fig, ax = plt.subplots()
CF = ax.contourf(X,Y,Z)
CS = ax.contour(X,Y,Z,colors='k')
lvls=CF.levels

ax.set_xlabel(r'$\xi$',fontsize=16)
ax.set_ylabel(r'$\beta$',fontsize=16)
ax.set_title(r'$R$ (bosons) for $\eta =$ {} '.format(eta))


plt.colorbar(CF)
plt.savefig('Rbosons-eta{}.png'.format(eta), bbox_inches = "tight",dpi=900)


plt.show()
```
