```python
from IGQG import *
from numpy import array,arange,meshgrid,concatenate, ones

from matplotlib import pyplot as plt
```

This notebook makes the countour graphs for curvature bosons presented in  Fig. 6 in the manuscript.

The commented cells  [9] obtain the graphs for $\eta=2$

```python
eta = power(2,-1)
```

Bosons

```python
x=concatenate([arange(.01,.9,.01),arange(.9,.99,.001),arange(.99,.999,.0002)])
beta = concatenate([arange(0,.03,.001),arange(.03,.2005,.005)])
```

```python
#eta=2
#beta = concatenate([beta,arange(.2,.601,.01)])
```

```python
aarray = array([A(eta,xx) for xx in x])
barray = array([B(eta,xx) for xx in x])

acarray = array([Ac(eta,xx) for xx in x])
bcarray = array([Bc(eta,xx) for xx in x])
```

```python
X,Y = meshgrid(x,beta)
Z = (.5*(Y**(eta+1)))*bose_Curvature(aarray,acarray,barray,bcarray,Y,eta) #the function bose_Curvature only does the unitless part
```

```python
fig, ax = plt.subplots()
CF = ax.contourf(X,Y,Z)
CS = ax.contour(X,Y,Z,colors='k')
lvls=CF.levels

ax.set_xlabel(r'$\xi$',fontsize=16)
ax.set_ylabel(r'$\beta$',fontsize=16)
ax.set_title(r'$R$ (bosons) for $\eta =$ {} '.format(eta))


plt.colorbar(CF)
plt.savefig('Rbosons-eta{}.png'.format(eta), bbox_inches = "tight",dpi=900)


plt.show()
```

```python
Z = (.5*(Y**(eta+1)))*bose_Curvature(aarray,zeros(aarray.shape),barray,zeros(barray.shape),Y,eta)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z,colors='k',levels=lvls)
CF = ax.contourf(X,Y,Z,levels=lvls,extend='max')
CF.cmap.set_over('#FFFF6F')

ax.set_xlabel(r'$\xi$',fontsize=16)
ax.set_ylabel(r'$\beta$',fontsize=16)
ax.set_title(r'$R_0$ (bosons) for $\eta =$ {} '.format(eta))

plt.colorbar(CF)
plt.savefig('RnCbosons-eta{}.png'.format(eta), bbox_inches = "tight",dpi=900)

plt.show()
```
