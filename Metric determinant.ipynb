```python
from IGQG import *
from numpy import array,arange,concatenate,zeros

from matplotlib import pyplot as plt
```

This notebook calculates the dimensionless quantities $\bar{g}_f$ and $\bar{g}_b$ which related to the metric determinant for fermions and bosons acording to equations (32) and (41) in the manuscript, respectively.

Fermions


```python
x = arange(.01,1,.01)
```

```python
etaarray = [.5,2]
fig, ax = plt.subplots()

for etaa in etaarray:
    y = fermi_Determinant(array([A(etaa,-xx) for xx in x]))
    line1, = ax.plot(x, y, label=('$\eta $= {}'.format(etaa)))

ax.set_xlim(0,1)

ax.set_xlabel(r'$\xi$', fontsize=15)
ax.set_ylabel(r'$\bar{g}_f(-\xi,\eta)$', fontsize=15)
ax.set_title(r'FD metric determinant ', fontsize=16)

plt.legend()
plt.savefig('Afermions.png',dpi=600)
plt.show()
```

Bosons

```python
eta = power(2,-1)
x=concatenate([arange(.01,.9,.01),arange(.9,.999,.001),arange(.999,1,.0001)])
```

```python
#eta=2
```

```python
aarray = array([A(eta,xx) for xx in x])
barray = array([B(eta,xx) for xx in x])

acarray = array([Ac(eta,xx) for xx in x])
bcarray = array([Bc(eta,xx) for xx in x])
```

```python
betas=[.05,.10,.15,.20]

fig, ax = plt.subplots()
y = bose_Determinant(aarray,zeros(aarray.shape),1,eta)
line1, = ax.plot(x, y,label='$g_0$')

for beta in betas:
    g = bose_Determinant(aarray,acarray,beta,eta)
    line1, = ax.plot(x, g,label=(r"${\beta}$ "+"= {}".format(beta)))

ax.set_ylim(0,200)
ax.set_xlim(0,1)

ax.set_xlabel(r'$\xi$', fontsize=15)
ax.set_ylabel(r'$\bar{g}_b(\xi,\eta)$', fontsize=15)
ax.set_title(r'BE metric determinant'+' for $\eta=${}'.format(eta), fontsize=18)
ax.legend(loc='upper left')#,bbox_to_anchor=(1.05, 1))
plt.savefig('gbosons-eta{}.png'.format(eta), bbox_inches = "tight",dpi=600)

plt.show()
```
